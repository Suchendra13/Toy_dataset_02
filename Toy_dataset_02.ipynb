## Importing the required libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.datasets import load_diabetes
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

## Load the dataset

```python
diabetes = load_diabetes()
X = diabetes.data
y = diabetes.target


print("--- Diabetes Dataset Loaded ---")
```

## Spliting Data of Training and Testing

```python
X_train, X_test, y_train, y_test = train_test_split(X, y , test_size = 0.2) #Training data of 80% and testing data of 20%.

print("--- Data split into training and testing sets ---")
```

## Training the model

```python
model = LinearRegression()
model.fit(X_train,y_train)

print("--- Linear Regression model trained ---")
```

```python
y_pred = model.predict(X_test)
```

## Evaluate the model's performance

```python
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test,y_pred)

print("\n--- Model Evaluation ---")
print(f"Mean Squared Error: {mse:.2f}")
print(f"R-squared (R²): {r2:.2f}")
```

## Prediction on new data

```python
new_patient_data = [diabetes.data[0]] 
prediction = model.predict(new_patient_data)

print("\n--- Example Prediction ---")
print(f"Data for new patient: {new_patient_data[0]}")
print(f"Predicted disease progression: {prediction.round(1)[0]}")
```

## Visualising the Results

```python
print("\n--- Generating Prediction vs. Actual Plot ---")
plt.figure(figsize=(8, 6))
plt.scatter(y_test, y_pred, color='blue', edgecolor='k', alpha=0.7, label='Predicted Values')
plt.plot([y.min(), y.max()], [y.min(), y.max()], 'k--', lw=2, label='Perfect Prediction')
plt.xlabel("Actual Disease Progression")
plt.ylabel("Predicted Disease Progression")
plt.title("Actual vs. Predicted Disease Progression")
plt.legend()
plt.grid(True)
plt.show()
```
